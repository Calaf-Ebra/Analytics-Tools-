Data Wrangling-preparing


```python
#df.isna().sum()#to see if we have missing values
df['name'].unique() #for catogorical data
```

```python
#Data Formating Conversion


#df["year"]=df["year"].astype(int)
df.rename(columns={"name":"nami"},inplace=True)#to cahnge column name
```

```python
#turn catogorical to numeric
pd.get_dummies(df['name'])
```

```python
#Missing Values

print(f" year has na? {pd.isnull(df['year']).values.any()}")
```

```python
med=df['year'].median()
df['year']=df['year'].fillna(med)
```

```python
pd.isnull(df['year']).values.any()
```

```python
# filling with most common class
df_clean = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
df_clean
```

```python
#Data Normalization.... Making values consistant.. no outlires..There are 3 equations to do so
#1-maximum value method
df["weight"]=df["weight"]/df["weight"].max()
```

```python
#2 Z scores Method
df["weight"]=(df["weight"]-df["weight"]).mean()/df["weight"].std()
```

```python
#Binning Data in Pandas... E.g price to low, m, high
bins=np.linspace(min(df["weight"]), max(df["weight"],4))
names=["Slim","Fat", "over"]
df["weight"]=pd.cut(df["weight"],bins,labels= names,include_lowest=True)
```

```python
#Box Plots
sns.boxplot(x="year", y="mpg",data=df)
```

```python
#Scatter Plots
y=df["mpg"]
x=df["year"]
plt.scatter(x,y, c= "green")
plt.title("Calaf")
plt.xlabel("year")
plt.ylabel("mpg")
```

```python
#Data Grouping is used to know realtionship between two variables.. predictor is the index
df-test=df[["mpg","year","horsepower"]]
df-grp=df-test.groupby(["mpg","year"] as-index=False).mean()
df-pivot=df-grp.pivot(index="mpg",columns="year")
```

```python
#heatmap
plt.pcolor(df-pivot,cmap='RdBa')
plt.colorbar()
plt.show()
```

```python
#Correlation
sns.regplot(x="mpg",y="year", data=df)
plt.ylim(0,)
```

```python
#pearson correlation... 1 good coor, -1 negative, 0 no correlation..P value how confident we are
pc=stats.pearsonr(df["year"], df["mpg"])
print(pc)
```
