**Chapter 3-Bar plots and Count plots**

```python
sns.countplot(x=df.Winner)
```

```python
sns.catplot(x=df.total_bill, y=df.day, kind="bar")
```

```python
sns.catplot(x=df.time,y=df.total_bill,kind="box",order=['Dinner','Lunch'])

plt.show()
```

```python
Changing the whiskers using
`
whis
`
By default, the whiskers extend to 1.5 * the interquartile range
Make them extend to 2.0 * IQR: whis=2.0
Show the 5th and 95th percentiles: whis=[5, 95]
Show min and max values: whis=[0, 100]
```

```python
sns.catplot(x=df.time,y=df.total_bill,kind="box",order=['Dinner','Lunch'],whis=[0,100])

plt.show()
```

```python
sns.catplot(x=df.time,y=df.total_bill,kind="pinit",order=['Dinner','Lunch'])

plt.show()

#point shows mean of qunatitive data..
#you can also add hue
#you can also set estimator= median for example
#change ci levl.. capsize=0.2 for example
```
